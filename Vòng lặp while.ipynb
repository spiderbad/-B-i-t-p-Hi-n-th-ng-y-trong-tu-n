```python
# Nhập vào 6 số a, b ,c, d, e, f Cho biết tổng 6 số là chẵn hay lẻ
a = int(input("Nhập số thứ 1: "))
b = int(input("Nhập số thứ 2: "))
c = int(input("Nhập số thứ 3: "))
d = int(input("Nhập số thứ 4: "))
e = int(input("Nhập số thứ 5: "))
f = int(input("Nhập số thứ 6: "))
total = a + b + c + d + e + f
if total%2 == 0:
    print("chẵn")
else:
    print("lẻ")
```

## Cấu trúc lặp while

```python
i = 0 
tong = 0
while i < 6:
    value = int(input("Nhập số: "))
    tong += value
    i += 1
print(f"Tổng của 6 số đã nhập là: {tong}")
```

### Ví dụ: In ra tất cả số chẵn từ 0 đến n

```python
x = 0
n = int(input("Nhập n: "))
while x <= n:
    if not x % 2:
        print(x)
    x += 1
```

### Ví dụ: In ra tất cả số lẻ từ n về 1

```python
n = int(input("Nhập n: "))
i = n
while i >= 1:
    if i % 2:
        print(i)
    i -= 1
```

```python
string = input("")
# n = len(string)
# x = string[-1:-n-1:-1]
# print(x)

i = len(string) - 1
while i >= 0:
    print(string[i], end ="")
    i -= 1
    
```

### Lệnh Break

Cú pháp:

`break`

Lệnh break sẽ giúp hủy vòng lặp đang thực hiện

##### Lưu ý:

##### - Lệnh break <b>BẮT BUỘC</b> phải được đặt trong vòng lặp
##### - Lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lắp lồng vào nhau)

#### Ví dụ

Sử dụng lệnh `while (True)` và break và in ra 10 lần câu "Hello mọi người"

```python
# Nhập n. In ra 3 số lẻ đầu tiên trong khoảng từ n đến 1 
n = int(input("Nhập n: "))
i = n 
count = 0 
while i >= 1:
    if i % 2 :
        print(i)
        count += 1 
    if count == 3:
        break
        
    i -= 1

        
```

## Lệnh continue

```python
# In ra tất cả những số từ 1 đến 10 trừ các số 3 , 4 , 6
i = 0 
while i < 10:
    i += 1
    if i ==3 or i == 4 or i == 6:
        continue
    print(i)
    
    
```
